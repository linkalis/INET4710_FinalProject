## Overview

This project explores non-negative matrix factorization (NMF) as a tool for "learning" land use classifications in aerial imagery.  Non-negative matrix factorization, which has been popularized in the context of recommender systems, is also proving to be a useful method in text mining and other diverse fields.  In general, NMF can be seen as an appealing technique for a class of problems that seek to decompose complex matrix-like data into a small number of latent "factors" that are reasonably interpretable by human intuition.  The non-negativity constraint means that the resulting factors can be interpreted as a set of factors/classes (e.g. product types, text genres, land use categories) and a set of weights linking each factor/class to the original observations (e.g. customers, documents, image pixels).  Several authors have noted the connection between non-negative matrix factorization algorithms and "spectral unmixing" problems for aerial imagery classification (see: Gillis, 2014; Bioucas-Dias, et al., 2012).  Indeed, NMF appears to be a technique that is solidly establishing itself within the remote sensing community in recent years.

The data I will focus on for this project is from a the [National Agricultural Imagery Program (NAIP)](https://www.fsa.usda.gov/programs-and-services/aerial-photography/imagery-programs/naip-imagery/), which is a large-scale goverment-administered data collection effort that makes orthorectified aerial photography available to government agencies and the general public on a regularly-updated basis.  This imagery is captured every 3 to 5 years, and it is captured during the agricultural growing season for most regions of the continental U.S.  This makes NAIP data a wonderful opportunity for analyzing land use in both agricultural and urban built environments over time.

The goal of this project is to establish whether non-negative matrix factorization techniques present a promising approach towards unsupervised land use classification in NAIP imagery.  I will also try to establish some potential pitfalls/issues that remain to be addressed when employing NMF techniques against NAIP imagery.  I will look at two different implementations of non-negative matrix factorization algorithms: the [NMF implementationin scikit-learn](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.NMF.html), and the [ALS implementation in Apache Spark](https://spark.apache.org/docs/2.2.0/ml-collaborative-filtering.html) (accompanied by a non-negativity constraint).  This project will focus on the following questions:

* Can non-negative matrix factorization algorithms "learn" land use classifications from 4-band aerial imagery?

* Are there any simple noise reduction techniques that can help improve the accuracy of the classifications "learned" from the data?

* Are there noticeable performance differences for different implementations of NMF algorithms that might make one algorithm or library preferable to another?

As a basis for comparison, I will also look at [land use classifications that were generated by the University of Minnesota's Remote Sensing and Geospatial Analysis Laboratory](https://gisdata.mn.gov/dataset/base-landcover-twincities) for the region of interest around the Twin Cities metro area.  The team who generated these land use classifications started from the same set of NAIP imagery I will explore here, but enriched it with ancillary datasets and used a different classification approach for establishing land use classifications.  The Remote Sensing and Geospatial Analysis Laboratory group employed a supervised object-based image analysis (OBIA) technique that required the creation of customized rule sets for training land use classification tasks.  This project does not attempt to replicate their classifications exactly, but rather seeks to assess how an unupervised NMF approach might measure up to these kinds of more labor-intensive supervised workflows for land use classification.  If the NMF approach explored here is able to learn even one or two of the land use categories accurately and efficiently, this is an important "win" for unsupervised approaches in land use classification problems.

Note that the filename structure for NAIP images is as follows: `S_DDdddQQ_qq_ZZ_R_YYYYMMDD_yyyymmdd.tif`

* S = Sensor Type
* DDddd = SE Reference point of 1 Degree Block (Latitude = DD, Longitude = ddd)
* QQ = Quad location (range 1 -64)
* qq = Quadrant indicator (NE, NW, SE, SW)
* ZZ = UTM Zone
* R = Resolution in meters
* YYYYMMDD = Acquisition date (year, month, day)
* yyyymmdd = Date NAIP JPG2000 data was created (year, month, day)

To learn a little more about the train and test images, we can plot them in the context of the entire set of NAIP images available across the state of Minnesota in the 2015 dataset.  To start, we can extract the bounding boxes from each of the images and store this information to a geodataframe.  (Note: A geodataframe--part of the 'geopandas' library--is essentially a "regular" pandas datafram that carries the added advantage of allowing you to store geometry columns alongside regular data columns.)

```python
def extract_img_bounding_boxes(img_list):
    ''' Loop through each image, and store information about them to a geodataframe.
    Convert to a geodataframe with the bounding boxes as the geometry column.
    See: https://gis.stackexchange.com/questions/285336/convert-polygon-bounding-box-to-geodataframe?rq=1
    '''

    rows_list = []

    for img_filepath in img_list:
        img = rasterio.open(img_filepath)
        row_dict = {}
        row_dict['filename'] = img.name
        row_dict['crs'] = str(img.crs)
        row_dict['bands'] = img.count
        row_dict['width'] = img.width
        row_dict['height'] = img.height
        row_dict['bounding_box'] = box(img.bounds[0], img.bounds[1], img.bounds[2], img.bounds[3])
        rows_list.append(row_dict)

    bounding_boxes_gdf = gpd.GeoDataFrame(pd.DataFrame(rows_list), geometry = 'bounding_box')
    bounding_boxes_gdf.crs = {'init' :'epsg:26915'}
    
    return(bounding_boxes_gdf)
```

### Visualize raster bands

We can start by plotting one of the NAIP images--the one we'll be using as the primary training image--and separate it out into its red, green, blue, and infrared bands.  This particular image represents the area near the St. Paul campus of the University of Minnesota.  It is a good choice to use as a training image because it contains many different landcover classes--including urban built features, water features, and even some agricultural area near the University's research plots.

For raster plotting and other raster manipulation tasks, we'll use a library called 'rasterio', which is a popular raster management library for Python that is maintained by MapBox.

```python
def plot_four_bands(img_filepath):
    ''' Given a filepath to a four-band GeoTIFF image, read in the image using rasterio, then
     visualize each of the four bands as separate plots in a matplotlib figures.
    '''
    
    img = rasterio.open(img_filepath)
    
    fig, (axr, axg, axb, axi) = plt.subplots(1, 4, figsize=(21,7))
    show((img, 1), ax=axr, cmap='Reds', title='red channel')
    show((img, 2), ax=axg, cmap='Greens', title='green channel')
    show((img, 3), ax=axb, cmap='Blues', title='blue channel')
    show((img, 4), ax=axi, cmap='Greys', title='infrared channel')
    
    img.close()
```

```python
plot_four_bands(train_images[0])
```

Next, we can look at all four of these bands together, displayed as a standard RGB image, and then plot the distribution of values observed for each band on a histogram.  When we do this, we see that all four of the bands have values that fall in the range of 0 - 255; this should not be surprising, as it is a standard convention for digital image formats to have band values ranging from 0 - 255.  Interestingly, there is a lot of overlap in the mid-range of values (~100 - 150 DN) on the green (#2) and infrared (#4) bands; this could represent foliage such as grass that is green-ish in color, and that also tends to reflect the infrared spectrum with mid-range intensity.  There is also a "hump" in the high end of the infrared band (~175 - 190 DN), suggesting a large number pixels--likely also representing foliage--that weigh quite high in infrared reflectance.

```python
def show_rgb_and_histogram(img_filepath):
    ''' Given a filepath to a four-band GeoTIFF image, read in the image using rasterio, then
     visualize the RGB version of the image alongside a histogram showing the distribution of
     values for each band.
    '''
    
    img = rasterio.open(img_filepath)
    
    fig, (axrgb, axhist) = plt.subplots(1, 2, figsize=(10, 5))
    show(img.read([1, 2, 3]), ax=axrgb)
    show_hist(img, bins=50, histtype='stepfilled', lw=0.0, stacked=False, alpha=0.3, ax=axhist)
    plt.show()
    
    img.close()
```

### Visualize NDVI

One final calculation and visualization we can perform is looking at the NDVI value for the image pixels.  The Normalized Difference Vegetation Index (NDVI) is a simple calculation based on the idea that plants and other types of foliage absorb red light, but reflect light in the near infrared (NIR) spectrum.  This yields the following calculation, whose values range from -1 to 1: 

$$NDVI = \dfrac{NIR - Red}{NIR + Red}$$

This very simple ratio helps discern between things that are likely to be plants (where NDVI values are > 0) and things that are likely to be built environments (where NDVI values are < 0).  Water is unique in that its NDVI value should be highly negative; a quick comparison of the images in this NAIP dataset suggests that a value <= -0.3 is likely to represent water.  Interestingly, NDVI also has a lesser-known cousin--the Normalized Difference Water Index (NDWI)--that has been successfully used in swimming pool detection in urban/suburban environments (See: [McFeeters, 2013](https://www.mdpi.com/2072-4292/5/7/3544)).  This suggests that NDVI could be a promising separation metric to explore before launching into our pixel classification task.

```python
def visualize_ndvi(img_filepath):
    
    np.seterr(divide='ignore', invalid='ignore')
    
    img = rasterio.open(img_filepath)
    band_r = img.read(1).astype(float)
    band_ir = img.read(4).astype(float)
    
    ndvi = np.empty(img.shape, dtype=rasterio.float32)
    ndvi = (band_ir - band_r) / (band_ir + band_r)

    ndvi_neg1 = np.empty(img.shape, dtype=rasterio.float32)
    ndvi_neg1 = np.where(np.logical_and(np.greater_equal((band_ir - band_r) / (band_ir + band_r), -0.3),
                                        np.less_equal((band_ir - band_r) / (band_ir + band_r), 0)), 1, 0)
    
    ndvi_neg2 = np.empty(img.shape, dtype=rasterio.float32)
    ndvi_neg2 = np.where(((band_ir - band_r) / (band_ir + band_r)) < -0.3, 1, 0)
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(21,7))
    
    show_hist(ndvi, bins=50, ax=ax1)
    
    ax2.imshow(ndvi, cmap='viridis')
    ax2.set_title('NDVI +/-')
    
    ax3.imshow(ndvi_neg1)
    ax3.set_title('-0.3 < NDVI < 0 (built environment)')
    
    ax4.imshow(ndvi_neg2)
    ax4.set_title('NDVI < -0.3 (water)')
```

Indeed, when we visualize the training image, separating out pixels whose NDVI values that lie between -0.3 and 0, and values that are less than -0.3, we see that it appears to provide a reasonably good way to discern the difference between water and urban built spaces:

```python
visualize_ndvi(train_images[0])
```

## Explore "Ground Truth" Comparison Data

Now, we can take a look at the comparison dataset: the [land cover classifications](https://gisdata.mn.gov/dataset/base-landcover-twincities) generated by the University of Minnesota's Remote Sensing and Geospatial Analysis Laboratory.  We download the dataset as a .GeoTIFF file.  The dataset itself contains a larger swath of the metro region that we are interested in for this analysis; it has been merged to extend to the 7-county boundaries of the entire Twin Cities metro area.  To allow for good comparisons between this dataset and the NAIP training images, we need to clip the comparison raster into four different chunks, each representing the coverage are of one of the original NAIP training images.

```python
def get_comparison_clip(gdf, row_nbr, comparison_img_filepath, output_filepath):
    ''' Takes a geodataframe containing polygon bounding boxes, a row number indicating which 
    image to use as the 'cookie cutter' whose boundaries to clip, and the filepath to a 
    comparison raster we wish to clip.  Generates a clip of the comparison raster that matches
    the gounding box inputs, then saves the result to a user-selected output filepath.
    '''
    
    img = rasterio.open(comparison_img_filepath)
    
    coords_to_clip = [json.loads(gdf.to_json())['features'][row_nbr]['geometry']]
    print(coords_to_clip)
    
    comparison_clip, clip_out_transform = mask(img, shapes=coords_to_clip, crop=True)
    print(comparison_clip.shape)
    
    
    out_tif = output_filepath + "comparison_clip" + str(row_nbr) + '.tif'
    out_meta = img.meta.copy() # get the metadata settings from the original comparison image
    epsg_code = int(img.crs.data['init'][5:])
    print(out_meta)
    
    out_meta.update({
        'driver': 'GTiff',
        'height': comparison_clip.shape[1], # substitute new size information
        'width': comparison_clip.shape[2],
        'transform': clip_out_transform
    })

    print(out_meta)
    
    with rasterio.open(out_tif, "w", **out_meta) as dest:
        dest.write(comparison_clip)
```

We can visualize the classifications present in the comparison clip:

```python
def visualize_comparison_classes(comparison_clip_filepath):
    ''' Visualize a single-band image featuring the classification scheme described here:
    https://conservancy.umn.edu/bitstream/handle/11299/181532/Classification%20scheme%20for%20Level%201%20and%202%20land%20cover%20classes.pdf?sequence=16&isAllowed=y
    See also: https://stackoverflow.com/questions/42496357/discrete-color-bar-with-tick-labels-in-between-colors
    '''
    
    img = rasterio.open(comparison_clip_filepath)
    classification_band = img.read(1) # The image should contain a single band with numeric classifications
    
    fig = plt.figure(figsize=(10, 6))

    classification_labels = [
        'none', 'grass', 'soil', 'buildings', 'roads_pavement', 'lakes_ponds', 
        'deciduous_trees', 'coniferous_trees', 'agriculture', 'emergent_wetland', 
        'forest_shrub_wetland' 
        #'river', 'extraction'
    ]

    image = plt.imshow(classification_band, cmap='Set3')
    cbar = plt.colorbar(image, ticks=range(12), label='classification')
    tick_locs = (np.arange(13) + 0.5)*(12-1)/12
    cbar.set_ticks(tick_locs)
    cbar.set_ticklabels(classification_labels)
```

We can also compute some summary statistics to determine what percent of the image each landcover class represents. When visualized in this way, it is easy to see, for example, that class 4 represents roads and class 5 represents lakes.  We also notice that most of the classes are represented in this image, making this a good candidate to use as a training image.  Note that if we are missing a class altogether in our training data, it will be impossible for any algorithm we employ to "learn" to identify that class using an unsupervised approach.

```python
def summarize_landcover_classes(classification_raster_filepath, plots_per_row):
    ''' Takes an image band that contains numeric classifications, and calculates the percentage
    of the total image area that is represented by that band. Plots an image of each band, 
    and returns a summary dataframe.
    '''

    img = rasterio.open(classification_raster_filepath)
    classification_band = img.read(1) # The image should contain a single band with numeric classifications
    
    class_nbr_list = []
    count_of_pixels_list = []
    pct_of_pixels_list = []
    
    total_pixels = classification_band.size
    classifications = np.unique(classification_band)
    classification_count = len(classifications)
    
    fig_rows = math.ceil(classification_count / plots_per_row)
    fig_cols = plots_per_row
    fig = plt.figure(figsize=(15, 15))
    
    for class_nbr in classifications:
        class_matrix = np.isin(classification_band, class_nbr)
        class_count = np.sum(class_matrix)
        class_pct = round((class_count / total_pixels) * 100, 3)
        
        # Append summary stats to summary lists
        class_nbr_list.append(class_nbr)
        count_of_pixels_list.append(class_count)
        pct_of_pixels_list.append(class_pct)
        
        label = str("Class " + str(class_nbr) + ": " + str(class_pct) + "%")
        plt.subplot(fig_rows, fig_cols, class_nbr + 1)
        plt.title(label)
        plt.imshow(class_matrix, cmap='binary')
    
    plt.show()
    
    summary_dict = { 
        "class_nbr" : class_nbr_list, 
        "count_of_pixels" : count_of_pixels_list, 
        "pct_of_pixels" : pct_of_pixels_list
    }
    
    img.close()
    
    return(pd.DataFrame.from_dict(summary_dict))
```
